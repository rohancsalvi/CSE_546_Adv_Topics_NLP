Just sentences

```python
def get_sentence(ent_token,file_name,chr_offset):
  filename = './sentences/' + file_name
  count= 0
  with open(filename, 'r') as file:
    while True:
      count += 1

    # Get next line from file
      line = file.readline()

    # if line is empty
    # end of file is reached
      if not line:
        break
      line = line.strip()
      line = line.replace("U . S .", "U.S.")
      line = line.replace("U . N .", "U.N.")
      if ent_token in line:
        return line
```

```python
from transformers import BertTokenizer
tz = BertTokenizer.from_pretrained("bert-base-cased")

```

```python
ent_train_list = df_train['Entity Token'].tolist()
file_train_list = df_train['File name'].tolist()
train_mention_mask = []
train_sent_encoding = []
for i in range(0,len(ent_train_list)):
  ent_token = ent_train_list[i]
  file_name = file_train_list[i].split(':')[0]
  chr_offset = file_train_list[i].split(':')[1]
  sentence = get_sentence(ent_token,file_name,chr_offset)
  #print(ent_token)
  if sentence is not None:
    tokens = tz.tokenize(sentence)
  #print(tokens)
  mention_mask = [0]* len(tokens)
  if ent_token in tokens:
    for i in range(0,len(tokens)):
      if tokens[i] == ent_token:
        mention_mask[i] = 1
  else:
    temp_list = []
    flag = False
    for i in range(0,len(tokens)):
      if flag == True:
        samp_str = ''
        for j in range(0,len(temp_list)):
          samp_str = samp_str + tokens[temp_list[j]]
        next_token = tokens[i]
        next_token = next_token.replace('##','')
        samp_str = samp_str + next_token
        if samp_str in ent_token:
          temp_list.append(i)
          if samp_str == ent_token:
            break
        else:
          flag = False
          temp_list = []
      if tokens[i] in ent_token and flag == False:
        flag = True
        temp_list.append(i)

    for i in range(0,len(tokens)):
      if i in temp_list:
        mention_mask[i] = (1/len(temp_list))
  train_mention_mask.append(mention_mask)
  train_sent_encoding.append(tz.convert_tokens_to_ids(tokens))
```

```python
from torch import nn
from transformers import BertModel

NUM_CATEGORIES = 210

class BertClassifier(nn.Module):

    def __init__(self, dropout=0.5):

        super(BertClassifier, self).__init__()

        self.bert = BertModel.from_pretrained('bert-base-cased')
        self.dropout = nn.Dropout(dropout)
        self.linear = nn.Linear(768, NUM_CATEGORIES)
        self.relu = nn.ReLU()
        self.sigmoid =nn.Sigmoid()

    def forward(self, input_id, mask,mention_mask,context_mask):

        _, pooled_output = self.bert(input_ids= input_id, attention_mask=mask,return_dict=False)
        dropout_output = self.dropout(pooled_output)

       # Average mention embedding
        mention_mask = ((1 - mention_mask) * -1e14).softmax(-1)
        #print(mention_mask.shape)
        #print(dropout_output.shape)
        mention_repr = (dropout_output * mention_mask).sum(1)
        mention_repr_v2 = (dropout_output * mention_mask)
        #print(mention_repr.shape)
        #print(mention_repr_v2.shape)
      #   # # Average context embedding
      #   # context_mask = ((1 - context_mask) * -1e14).softmax(-1).unsqueeze(-1)
      #   # context_repr = (dropout_output * context_mask).sum(1)

      #   # # Concatenate mention and context representations
      #   #combine_repr = torch.cat([mention_repr, context_repr], dim=1)

        scores = self.linear(mention_repr_v2)


        ##linear_output = self.linear(dropout_output)
        ##final_scores = self.sigmoid(scores)

        return scores
```

```python
from torch.optim import Adam
import numpy as np

def train(model, learning_rate, epochs):

    use_cuda = torch.cuda.is_available()
    device = torch.device("cuda" if use_cuda else "cpu")

    criterion = nn.MultiLabelSoftMarginLoss()
    optimizer = Adam(model.parameters(), lr= learning_rate)
    batch_size = 20
    if use_cuda:

            model = model.cuda()
            criterion = criterion.cuda()

    for epoch_num in range(epochs):

            print(epoch_num)
            for i in range(0,len(train_sent_encoding), batch_size):

                train_label = torch.from_numpy(transformed_train_targets[i : i + batch_size]).to(device)
                mask = torch.from_numpy(np.array(train_context_vector[i: i + batch_size], dtype=np.int)).to(device)
                input_id = torch.from_numpy(np.array(train_sent_encoding[i: i + batch_size], dtype=np.int)).to(device)
                men_mask = torch.from_numpy(np.array(train_mention_mask[i: i + batch_size], dtype=np.int)).to(device)

                #print(input_id.shape)

                output = model(input_id, mask,men_mask,mask)

                loss = criterion(output, train_label)

                model.zero_grad()
                loss.backward()
                optimizer.step()
                if(i%500 == 0):
                  print(i)
                  print(loss)



EPOCHS = 4
model = BertClassifier()
LR = 1e-6

train(model, LR, EPOCHS)
```
